```python
import random
import os
import pandas as pd
dataset_path = "../ASSIGNMENT-1/telugu_dataset.txt" 
NLP_LAB= "./" 
```

```python
with open(dataset_path, "r", encoding="utf-8") as f:
    sentences = [line.strip() for line in f if line.strip()]

print(f"Total sentences in dataset: {len(sentences)}")
```

```python

random.seed(42)  # for reproducibility
random.shuffle(sentences)

val_set = sentences[:1000]
test_set = sentences[1000:2000]
train_set = sentences[2000:]

print(f"Training set: {len(train_set)}")
print(f"Validation set: {len(val_set)}")
print(f"Test set: {len(test_set)}")
```

```python
os.makedirs(NLP_LAB, exist_ok=True)
with open(os.path.join(NLP_LAB, "train.txt"), "w", encoding="utf-8") as f:
    for line in train_set:
        f.write(line + "\n")

with open(os.path.join(NLP_LAB, "validation.txt"), "w", encoding="utf-8") as f:
    for line in val_set:
        f.write(line + "\n")

with open(os.path.join(NLP_LAB, "test.txt"), "w", encoding="utf-8") as f:
    for line in test_set:
        f.write(line + "\n")

print("Data splits saved successfully!")
```

```python
val_df = pd.DataFrame({"sentence": val_set})
test_df = pd.DataFrame({"sentence": test_set})

val_df.to_csv(os.path.join(NLP_LAB, "validation.csv"), index=False, encoding="utf-8")
test_df.to_csv(os.path.join(NLP_LAB, "test.csv"), index=False, encoding="utf-8")

print("Data splits saved successfully as TXT and CSV!")
```
