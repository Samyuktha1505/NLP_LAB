```python
from datasets import load_dataset

dataset = load_dataset("ai4bharat/IndicCorpV2", name="indiccorp_v2", split="tel_Telu",streaming=True)
```

```python
import re

pattern = re.compile(r'(?<=[.!?])\s+')

all_sentences = []
num = 0

for i in dataset:
    if i['text'].strip():
        num += 1
        print(i['text'])
        print('\n')

        telugu_sentences = pattern.split(i['text'])
        all_sentences.extend(telugu_sentences)

    if num == 5:
        break

joined_text = '\n'.join(all_sentences)

no_of_sentences = len(all_sentences)

print("Number of sentences:", no_of_sentences)
```

```python
import re

pattern = re.compile(
    r'https?://\S+|'                     # URLs
    r'\b[\w\.-]+@[\w\.-]+\.\w{2,4}\b|'   # email addresses
    r'\d{1,2}[/-]\d{1,2}[/-]\d{2,4}|'    # dates like 12-05-2020
    r'\d+\.\d+|\d+|'                     # numbers including decimals
    r'[\u0C00-\u0C7F]+|'                 # Telugu words
    r'[^\s\w\u0C00-\u0C7F]'              # punctuation
)

all_tokens = []  # store all tokens here
num = 0

for i in dataset:
    if i['text'].strip():
        num += 1
        print(i['text'])
        print('\n')

        tokens = pattern.findall(i['text'])
        all_tokens.extend(tokens)  # add to the global list

    if num == 5:
        break

unique_tokens = len(set(all_tokens))
no_of_words = len(all_tokens)

print("Tokens:", all_tokens)
print("Unique tokens:", unique_tokens)
print("Total words/tokens:", no_of_words)
```

```python
import re

pattern = re.compile(r'[\u0C00-\u0C7F0-9.,!?;:"\'()\[\]{}\-—–…\s]')

all_chars = []
num = 0

for i in dataset:
    if i['text'].strip():
        num += 1
        chars = pattern.findall(i['text'])
        all_chars.extend(chars)
    if num == 5:
        break
no_of_chars = len(all_chars)

print("Characters:", all_chars)
print("Number of characters:", no_of_chars)
```
