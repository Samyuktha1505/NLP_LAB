```python
from automata.fa.dfa import DFA
from visual_automata.fa.dfa import VisualDFA
```

```python
import string
lower = list(string.ascii_lowercase)
upper = list(string.ascii_uppercase)
input_symbols = set(lower + upper)

all_ascii = set(chr(i) for i in range(32, 127))

invalid_inputs = all_ascii - input_symbols
```

```python
transitions = {
    'q0': {},
    'q1': {},
    'q2': {},
    'qReject': {}
}

for c in lower + upper:
    transitions['q0'][c] = 'q1'

for c in lower:
    transitions['q1'][c] = 'q2'

for c in lower:
    transitions['q2'][c] = 'q2'

for state in ['q0', 'q1', 'q2']:
    for c in invalid_inputs:
        transitions[state][c] = 'qReject'

for c in upper:
    transitions['q1'][c] = 'qReject'
    transitions['q2'][c] = 'qReject'

transitions['qReject'] = {c: 'qReject' for c in all_ascii}
```

```python
dfa = VisualDFA(
    states={'q0', 'q1', 'q2', 'qReject'},
    input_symbols=all_ascii,
    transitions=transitions,
    initial_state='q0',
    final_states={'q1', 'q2'}
)
```
